# Loading Libraries and Dataset and Start Data Pre-Processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, roc_curve, roc_auc_score
```

```python
# Load your dataset (replace with your actual file path)
df_a = pd.read_csv('DAMO510, Module 08, Dataset1.csv')

# Preview the data
display(df_a.head())
print(df_a.shape)
```

```python
# Define target variable
y = df_a['segment']

# Select numerical features for LDA
X = df_a.drop(columns=["Cust_No", "zip_code", "segment"])
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
LDA = LinearDiscriminantAnalysis()
LDA.fit(X_train, y_train)
y_pred_lda = LDA.predict(X_test)
```

```python
soldf = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_lda})
print(soldf.head())
```

```python
confusion_matrix(y_test, y_pred_lda)
```

```python
print(classification_report(y_test, y_pred_lda))
```

```python
# suport vector machine
from sklearn.svm import LinearSVC
svm = LinearSVC()
svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred_svm)
```

```python
print(classification_report(y_test, y_pred_svm))
```
